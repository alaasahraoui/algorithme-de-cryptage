```python
#3 le carré de polybe 

def polybe_encreption():
    import numpy as np
    arr= np.array([['a', 'b', 'c','d','e'],
              ['f', 'g', 'h','ij','k'],
              ['l', 'm','n','o','p'],
              ['q','r','s','t','u'],
              ['v','w','x','y','z']
             ])
    text_clair=input("le text en clair =>")
    coded_text=""
    for char in text_clair:
        row=np.where(arr == char)[0]
        row_index=row[0]+1
        
        col=np.where(arr == char)[1]
        col_index=col[0]+1
        coded_char=str(row_index)+str(col_index)
        coded_text+=coded_char
    print('le text chiffré est:',coded_text)
    
def polybe_decryption():
    import numpy as np
    import re 
    arr= np.array([['a', 'b', 'c','d','e'],
              ['f', 'g', 'h','ij','k'],
              ['l', 'm','n','o','p'],
              ['q','r','s','t','u'],
              ['v','w','x','y','z']
             ])
    coded_text=input("le text chiffré =>")
    coded_text_array=re.findall('..',coded_text)
    text_clair=""
    for elm in coded_text_array:
        row_index=int(elm[0])-1
        
        col_index=int(elm[1])-1
        text_clair+= arr[row_index][col_index]
    print("le text en clair est =>",text_clair)
polybe_encreption()
polybe_decryption()     
```

```python
#véginaire algorithme 

def véginaire_encrypt():
    plaintext=input('le text en clair => (en majuscule uniquement)')
    key=input('la clé (en majuscule uniquement)=>')
    key_length = len(key)
    
    
    key_as_int = [ord(i) for i in key]
    plaintext_int = [ord(i) for i in plaintext]
   
   
    ciphertext = ''
    for i in range(len(plaintext_int)):
        value = (plaintext_int[i] + key_as_int[i % key_length]) % 26
        ciphertext += chr(value + 65)
        print(ciphertext)
    return ciphertext


def véginaire_dencrypt():
    ciphertext=input("text chiffré =>(en majuscule uniquement)")
    key=input("la clé (en majuscule uniquement)")
    key_length = len(key)
    key_as_int = [ord(i) for i in key]
    ciphertext_int = [ord(i) for i in ciphertext]
    plaintext = ''
    for i in range(len(ciphertext_int)):
        value = (ciphertext_int[i] - key_as_int[i % key_length]) % 26
        plaintext += chr(value + 65)
        print(plaintext)
    return plaintext


véginaire_encrypt()
véginaire_dencrypt()
```
